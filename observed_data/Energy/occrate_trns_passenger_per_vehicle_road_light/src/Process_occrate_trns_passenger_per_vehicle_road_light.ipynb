```python
# Add Parent Folder to System Path

import os
import sys
import inspect

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, parentdir) 

####

import pandas as pd
import numpy as np

#Variable name
var_name='occrate_trns_passenger_per_vehicle_road_light'
#DATAFRAMES
df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_data=pd.read_excel ('../row_data/2009 29 - Occupancy rates of passenger.xls',sheet_name='TERM29_BD_Car')
df_data
```

```python
df_data_copy=df_data
new_header = df_data.iloc[1] #grab the first row for the header
df_data_copy = df_data[3:] #take the data less the header row
df_data_copy.columns = new_header #set the header row as the df header
df_data_copy.fillna(0)
```

```python
#### HISTORICAL ######

## Direct calculation lead to a avrerage value of 1.673 passenger per vehicle,
#if light vehicles can carry 4 passengers then the average occupacy is 1.673/4=0.41825


countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)


years=np.arange(1995,2008,1)
variable_vect=np.zeros(len(years))+0.41825

coun_year=[]
for count in zip(countries,ISO3):
    for year,var in zip(years,variable_vect):
        coun_year.append([count[0],count[1],int(year),var])
        
df_historical_out = pd.DataFrame(coun_year, columns=['Country','ISO3', 'Year',var_name])
df_historical_out
```

```python
df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
```

```python
#### PROJECTED ######

#Prediction for the next years will be the last historical value
years2=years
variable_vect2=variable_vect
for y in np.arange(min(years2),2051,1):
    if y not in years2:
        years2=np.append(years2,y)
        variable_vect2=np.append(variable_vect2,variable_vect2[-1])

coun_year2=[]
for count in zip(countries,ISO3):
    for y_p in zip(years2,variable_vect2):
        coun_year2.append([count[0],count[1],y_p[0],y_p[1]])
coun_year2

df_projected_out = pd.DataFrame(coun_year2, columns=['Country','ISO3', 'Year',var_name])
df_projected_out
```

```python
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
```
