```python
# Add Parent Folder to System Path

import os
import sys
import inspect

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, parentdir) 

####

import pandas as pd
import numpy as np

#Price originally in usd/m^3

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
```

```python
df_countries
```

```python
#### HISTORICAL ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=np.arange(2000,2023,1)
ppb=np.zeros(len(years))+1
coun_year=[]
for count in zip(countries,ISO3):
    for year,pb in zip(years,ppb):
        coun_year.append([count[0],count[1],year,pb])
df_historical_out = pd.DataFrame(coun_year, columns=['Country','ISO3', 'Year','occrate_trns_passenger_per_vehicle_powered_bikes'])
df_historical_out.to_csv('../input_to_sisepuede/historical/occrate_trns_passenger_per_vehicle_powered_bikes.csv')
df_historical_out
```

```python
#### PROJECTED ######

#Prediction for the next years will be the last historical value
countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years2=np.arange(2000,2051,1)
ppb2=np.zeros(len(years2))+1
coun_year2=[]
for count in zip(countries,ISO3):
    for year,pb in zip(years2,ppb2):
        coun_year2.append([count[0],count[1],year,pb])

df_projected_out = pd.DataFrame(coun_year2, columns=['Country','ISO3', 'Year','occrate_trns_passenger_per_vehicle_powered_bikes'])
df_projected_out.to_csv('../input_to_sisepuede/projected/occrate_trns_passenger_per_vehicle_powered_bikes.csv')
df_projected_out
```
